```python
def inserir_dados(sql,dados):
    
    try:
      
        cursor = conn.cursor()
        cursor.executemany(sql,dados)
        conn.commit()

        if cursor.rowcount == -1:
            print("Número de registros não disponível com 'rowcount'. Tentando contar as inserções.")
            print(f"{len(dados)} registros inseridos com sucesso.")
            return dados
        else:
            print(f"{cursor.rowcount} registros inseridos com sucesso.")
        cursor.close()
    except KeyError:
        return "Chave dos valores errados"
    except Exception as error:
        return error
  
    
```

```python
def pegar_dados(sql):
    try:
        cursor = conn.cursor()
        
        cursor.execute(sql)
        
        dados = cursor.fetchall()

        if len(dados) > 0:
            print("Dados pegos com sucesso")
            return dados  
        else:
            print("Não foi encontrado nenhum dado.")
            return []  
        
    except KeyError:
        return "Chave dos valores errados"
    except Exception as error:
        return f"Erro ao pegar os dados: {error}"
    finally:
        cursor.close()
```

```python
def gerar_data_aleatoria(inicio_ano, fim_ano):
    data_inicio = datetime(inicio_ano, 1, 1)
    data_fim = datetime(fim_ano, 12, 31)
 
    delta = data_fim - data_inicio
    dias_aleatorios = random.randint(0, delta.days)

    data_aleatoria = data_inicio + timedelta(days=dias_aleatorios)
    return data_aleatoria.date()
```
